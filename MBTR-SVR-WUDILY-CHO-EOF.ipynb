```
from rdkit import Chem
from rdkit.Chem import Draw
import os

# Replace df['Canonical_SMILES'] with your actual list of SMILES
all_smiles = df['Canonical_SMILES']

batch_size = 50
num_batches = len(all_smiles) // batch_size

# Process full batches
for i in range(num_batches):
    start_idx = i * batch_size
    end_idx = (i + 1) * batch_size
    smiles_batch = all_smiles[start_idx:end_idx]

    mols = []

    for smile in smiles_batch:
        mol = Chem.MolFromSmiles(smile)
        mols.append(mol)

    img = Draw.MolsToGridImage(
        mols,
        molsPerRow=10,
        subImgSize=(200, 200),
        legends=[x for x in smiles_batch],
        returnPNG=False
    )

    os.makedirs('personal/MBTR-SVR-EOF/Species_Images', exist_ok=True)

    img.save(f'personal/MBTR-SVR-EOF/Species_Images/{start_idx}_{end_idx - 1}.png', dpi=(300, 300))

# Process remaining species if any
if len(all_smiles) % batch_size != 0:
    start_idx = num_batches * batch_size
    end_idx = len(all_smiles)
    remaining_smiles = all_smiles[start_idx:end_idx]

    remaining_mols = []

    for smile in remaining_smiles:
        mol = Chem.MolFromSmiles(smile)
        remaining_mols.append(mol)

    remaining_img = Draw.MolsToGridImage(
        remaining_mols,
        molsPerRow=10,
        subImgSize=(200, 200),
        legends=[x for x in remaining_smiles],
        returnPNG=False
    )

    remaining_img.save(f'personal/MBTR-SVR-EOF/Species_Images/{start_idx}_{end_idx - 1}.png', dpi=(300, 300))
```

```
from sklearn.model_selection import KFold
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error
import math


# Create empty lists to store results

train_rmse_scores = []
test_rmse_scores = []

# Create directories to save the data
os.makedirs('personal/MBTR-SVR-EOF/data', exist_ok=True)

kf = KFold(n_splits=10, random_state=random_seed, shuffle=True)
for fold, (train_idx, test_idx) in enumerate(kf.split(X_scaled, y), 1):
    X_train, X_test = X_scaled[train_idx], X_scaled[test_idx], 
    y_train, y_test = y[train_idx], y[test_idx] # Directly indexing NumPy arrays

    grid_search.best_estimator_.fit(X_train, y_train)

    y_pred_train = grid_search.best_estimator_.predict(X_train)
    y_pred_test = grid_search.best_estimator_.predict(X_test)
   
    train_rmse = math.sqrt(mean_squared_error(y_train, y_pred_train))
    test_rmse = math.sqrt(mean_squared_error(y_test, y_pred_test))

    train_rmse_scores.append(train_rmse)
    test_rmse_scores.append(test_rmse)
  
    print(f"Fold {fold}:")
    print("Train R2:", r2_score(y_train, y_pred_train))
    print('Train MAE:', mean_absolute_error(y_train, y_pred_train))
    print('Train MSE:', mean_squared_error(y_train, y_pred_train))
    print('Train RMSE:', math.sqrt(mean_squared_error(y_train, y_pred_train)))
    print('Train Chemical Accuracy:', 2 * math.sqrt(mean_squared_error(y_train, y_pred_train)))

    print("Test R2:", r2_score(y_test, y_pred_test))
    print('Test MAE:', mean_absolute_error(y_test, y_pred_test))
    print('Test MSE:', mean_squared_error(y_test, y_pred_test))
    print('Test RMSE:', math.sqrt(mean_squared_error(y_test, y_pred_test)))
    print('Test Chemical Accuracy:', 2 * math.sqrt(mean_squared_error(y_test, y_pred_test)))

    # Calculate average RMSE across all folds
    avg_train_rmse = np.mean(train_rmse_scores)
    avg_test_rmse = np.mean(test_rmse_scores)

    print("\nAverage Train RMSE across all folds:", avg_train_rmse)
    print("Average Test RMSE across all folds:", avg_test_rmse)
    print('Average Chemical Accuracy across all folds:', 2 * avg_test_rmse)

    # Save X_train and y_train to CSV files
    train_df = pd.DataFrame(X_train, columns=[f'feature_{i}' for i in range(X_train.shape[1])])
    train_df['target'] = y_train
    train_df.to_csv(f'personal/MBTR-SVR-EOF/data/fold_{fold}_X_train.csv', index=False)

    # Save X_test and y_test to CSV files
    test_df = pd.DataFrame(X_test, columns=[f'feature_{i}' for i in range(X_test.shape[1])])
    test_df['target'] = y_test
    test_df.to_csv(f'personal/MBTR-SVR-EOF/data/fold_{fold}_X_test.csv', index=False)

# Calculate absolute error for training and testing data
    abs_error_train = abs(y_train - y_pred_train)
    abs_error_test = abs(y_test - y_pred_test)
```

```
import os
import matplotlib.pyplot as plt

def plotVersusComparison(y_true, y_pred, plot_type):
    """
    Plot a comparison between true and predicted values.

    Parameters:
        y_true (array-like): True target values.
        y_pred (array-like): Predicted target values.
        plot_type (str): Type of plot ('test' or 'train').

    Returns:
        None
    """
    plt.figure(figsize=(10, 8), dpi=150)
    plt.plot(y_true, 'o-', color="#54B345", linewidth=0.7, markersize=3, label="True")
    plt.plot(y_pred, 'o-', color="#ffc34e", linewidth=0.7, markersize=3, label="Predicted")
    plt.legend(loc='best', prop={"size": 7})
    plt.title(f'Comparison - {plot_type.capitalize()} Set')
    plt.grid(True, linestyle='--', linewidth=0.5, color='gray')
    plt.tight_layout()
    plt.gca().set_facecolor('white')  # Setting background to white

    # Create directory if it doesn't exist
    save_dir = 'personal/MBTR-SVR-EOF/png/'
    os.makedirs(save_dir, exist_ok=True)

    # Save and show plot
    save_path = os.path.join(save_dir, f'Comparison_{plot_type}.png')
    plt.savefig(save_path, bbox_inches='tight', pad_inches=0.1)
    plt.show()

plotVersusComparison(y_test, y_pred1, 'test')
plotVersusComparison(y_train, y_pred0, 'train')
```

```
# Import necessary libraries
from chemplot import Plotter
from io import StringIO
import sys
import pandas as pd

# Assuming smiles_list is a list of SMILES strings

# Create a Plotter object from the SMILES list
cp = Plotter.from_smiles(smiles_list)

# Generate a UMAP (Uniform Manifold Approximation and Projection) plot
umap_output = StringIO()
sys.stdout = umap_output  # Redirect stdout to capture printed output
cp.umap()
umap_output.seek(0)  # Reset the StringIO object to the beginning

# Perform clustering on the data with 10 clusters
cluster_output = StringIO()
sys.stdout = cluster_output  # Redirect stdout to capture printed output
cp.cluster(n_clusters=10)
cluster_output.seek(0)  # Reset the StringIO object to the beginning

# Reset stdout to its original value
sys.stdout = sys.__stdout__

# Extract UMAP coordinates data
umap_lines = umap_output.getvalue().strip().split('\n')[1:]  # Exclude header
umap_data = [line.split() for line in umap_lines]
umap_df = pd.DataFrame(umap_data, columns=['UMAP-1', 'UMAP-2', 'clusters'])

# Extract cluster information
cluster_lines = cluster_output.getvalue().strip().split('\n')[1:]  # Exclude header
cluster_data = [line.split() for line in cluster_lines]
cluster_df = pd.DataFrame(cluster_data, columns=['Cluster', 'Count'])

# Write the DataFrames to an Excel file
with pd.ExcelWriter('personal/MBTR-SVR-EOF/dataset/umap_coordinates_and_clusters.xlsx') as writer:
    umap_df.to_excel(writer, sheet_name='UMAP', index=False)
    cluster_df.to_excel(writer, sheet_name='Clusters', index=False)
```
