```python
from rdkit import Chem
from rdkit.Chem import Draw
import os

# Replace df['Canonical_SMILES'] with your actual list of SMILES
all_smiles = data['Canonical_SMILES']

batch_size = 50
num_batches = len(all_smiles) // batch_size

# Process full batches
for i in range(num_batches):
    start_idx = i * batch_size
    end_idx = (i + 1) * batch_size
    smiles_batch = all_smiles[start_idx:end_idx]

    mols = []

    for smile in smiles_batch:
        mol = Chem.MolFromSmiles(smile)
        mols.append(mol)

    img = Draw.MolsToGridImage(
        mols,
        molsPerRow=10,
        subImgSize=(200, 200),
        legends=[x for x in smiles_batch],
        returnPNG=False
    )

    os.makedirs('personal/MRD-SVR-EOF/Species_Images', exist_ok=True)

    img.save(f'personal/MRD-SVR-EOF/Species_Images/{start_idx}_{end_idx - 1}.png', dpi=(300, 300))

# Process remaining species if any
if len(all_smiles) % batch_size != 0:
    start_idx = num_batches * batch_size
    end_idx = len(all_smiles)
    remaining_smiles = all_smiles[start_idx:end_idx]

    remaining_mols = []

    for smile in remaining_smiles:
        mol = Chem.MolFromSmiles(smile)
        remaining_mols.append(mol)

    remaining_img = Draw.MolsToGridImage(
        remaining_mols,
        molsPerRow=10,
        subImgSize=(200, 200),
        legends=[x for x in remaining_smiles],
        returnPNG=False
    )

    remaining_img.save(f'personal/MRD-SVR-EOF/Species_Images/{start_idx}_{end_idx - 1}.png', dpi=(300, 300))
```

```python
import os
import numpy as np
from sklearn.model_selection import KFold
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error
import math

# Resetting indices of X_scaled and y
X_scaled = pd.DataFrame(X_scaled).reset_index(drop=True)
y = pd.Series(y).reset_index(drop=True)

# Create empty lists to store results
train_rmse_scores = []
test_rmse_scores = []

# Create directories to save the data
os.makedirs('personal/MRD-SVR-EOF/data', exist_ok=True)

kf = KFold(n_splits=10, random_state=random_seed, shuffle=True)
for fold, (train_idx, test_idx) in enumerate(kf.split(X_scaled), 1):
    X_train, X_test = X_scaled.iloc[train_idx], X_scaled.iloc[test_idx]
    y_train, y_test = y.iloc[train_idx], y.iloc[test_idx]
    
    grid_search.best_estimator_.fit(X_train, y_train)

    y_pred_train = grid_search.best_estimator_.predict(X_train)
    y_pred_test = grid_search.best_estimator_.predict(X_test)
   
    train_rmse = math.sqrt(mean_squared_error(y_train, y_pred_train))
    test_rmse = math.sqrt(mean_squared_error(y_test, y_pred_test))

    train_rmse_scores.append(train_rmse)
    test_rmse_scores.append(test_rmse)

    # Create DataFrames for train and test sets
    train_data = pd.DataFrame({
        'Index': train_idx,
        'Reported Data': y_train,
        'Predicted Data': y_pred_train,
        'Absolute Error': abs(y_train - y_pred_train)
    })
    test_data = pd.DataFrame({
        'Index': test_idx,
        'Reported Data': y_test,
        'Predicted Data': y_pred_test,
        'Absolute Error': abs(y_test - y_pred_test)
    })

    # Save train and test data to Excel files
    train_data.to_excel(f'personal/MRD-SVR-EOF/data/fold_{fold}_train_data.xlsx', index=False)
    test_data.to_excel(f'personal/MRD-SVR-EOF/data/fold_{fold}_test_data.xlsx', index=False)

    print(f"Fold {fold}:")
    print("Train R2:", r2_score(y_train, y_pred_train))
    print('Train MAE:', mean_absolute_error(y_train, y_pred_train))
    print('Train MSE:', mean_squared_error(y_train, y_pred_train))
    print('Train RMSE:', math.sqrt(mean_squared_error(y_train, y_pred_train)))
    print('Train Chemical Accuracy:', 2 * math.sqrt(mean_squared_error(y_train, y_pred_train)))

    print("Test R2:", r2_score(y_test, y_pred_test))
    print('Test MAE:', mean_absolute_error(y_test, y_pred_test))
    print('Test MSE:', mean_squared_error(y_test, y_pred_test))
    print('Test RMSE:', math.sqrt(mean_squared_error(y_test, y_pred_test)))
    print('Test Chemical Accuracy:', 2 * math.sqrt(mean_squared_error(y_test, y_pred_test)))

    # Calculate average RMSE across all folds
    avg_train_rmse = np.mean(train_rmse_scores)
    avg_test_rmse = np.mean(test_rmse_scores)

    print("\nAverage Train RMSE across all folds:", avg_train_rmse)
    print("Average Test RMSE across all folds:", avg_test_rmse)
    print('Average Chemical Accuracy across all folds:', 2 * avg_test_rmse)

    # Save X_train and y_train to CSV files
    np.savetxt(f'personal/MRD-SVR-EOF/data/fold_{fold}_X_train.csv', X_train, delimiter=',')
    np.savetxt(f'personal/MRD-SVR-EOF/data/fold_{fold}_X_test.csv', X_test, delimiter=',')

    # Save y_train and y_test to CSV files
    np.savetxt(f'personal/MRD-SVR-EOF/data/fold_{fold}_y_train.csv', y_train, delimiter=',')
    np.savetxt(f'personal/MRD-SVR-EOF/data/fold_{fold}_y_test.csv', y_test, delimiter=',')
    

    # Calculate absolute error for training and testing data
    abs_error_train = abs(y_train - y_pred_train)
    abs_error_test = abs(y_test - y_pred_test)
```
